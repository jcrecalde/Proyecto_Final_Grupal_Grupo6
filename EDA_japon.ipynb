```python
import requests
import json
import pandas as pd 
import re  
import numpy as np
```

```python
url= 'https://www.jma.go.jp/bosai/quake/data/list.json'

response =requests.get(url)
data=response.json()
df_japon= pd.DataFrame(data)
```

```python
df_japon
```

```python
df_japon.isnull().sum()
```

Cambio el formato de la fecha de la informacion del sismo: rdt

```python
#Columna rdt
# Convertir la columna 'fecha' a un objeto datetime
df_japon['rdt'] = pd.to_datetime(df_japon['rdt'])

# Eliminar la zona horaria de la columna 'fecha'
df_japon['rdt'] = df_japon['rdt'].dt.tz_localize(None)

# Convertir la columna 'fecha' de vuelta a un formato de cadena si es necesario
df_japon['rdt'] = df_japon['rdt'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
#Columna at 
df_japon['at'] = pd.to_datetime(df_japon['at'])

# Eliminar la zona horaria de la columna 'fecha'
df_japon['at'] = df_japon['at'].dt.tz_localize(None)

# Convertir la columna 'fecha' de vuelta a un formato de cadena si es necesario
df_japon['at'] = df_japon['at'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df_japon
```

Dropeo columnas que no voy a utilizar

```python
#df_japon = df_japon.drop("ttl", axis = 1) 
#df_japon = df_japon.drop("anm", axis = 1) 
#df_japon = df_japon.drop("json", axis = 1) 
#df_japon = df_japon.drop("int", axis = 1) 
#df_japon = df_japon.drop("ctt", axis = 1) 
df_japon = df_japon.drop("cod", axis = 1) 
```

Me fijo que datos tiene ift que es la informacion del sismo

```python
df_japon["ift"].unique()
```

```python
#Cambio 発表 
df_japon["ift"] = df_japon["ift"].replace("発表", "anuncio")
```

Cambio de nombre a las columnas 

```python
df_japon = df_japon.rename(columns={"rdt" : "time"}) 
df_japon = df_japon.rename(columns={"eid" : "id"}) 
df_japon = df_japon.rename(columns={"ift" : "tipo_informacion"}) 
df_japon = df_japon.rename(columns={"ser" : "num_serie_sismo"}) 
df_japon = df_japon.rename(columns={"at" : "Update"}) 
df_japon = df_japon.rename(columns={"acd" : "Profundidad"})  
df_japon = df_japon.rename(columns={"maxi" : "intencidad_max"}) 
df_japon = df_japon.rename(columns={"en_ttl" : "title"}) 
df_japon = df_japon.rename(columns={"en_anm" : "nombre_area_epicentral"})
```

Extraigo latitud y longitud de cod

```python
print(df_japon['cod'])
```

```python
# Aplicar una expresión regular para extraer la latitud y longitud de la columna "cod"
df_japon[['latitud', 'longitud']] = df_japon['cod'].str.extract(r'([+-]?\d+\.\d+)([+-]?\d+\.\d+)')
```

```python
# Convertir la latitud y longitud al formato "-124.4023333"
df_japon['latitud'] = df_japon['latitud'].apply(lambda x: '{:.7f}'.format(-float(x)) if float(x) < 0 else '{:.7f}'.format(float(x)))
df_japon['longitud'] = df_japon['longitud'].apply(lambda x: '{:.7f}'.format(-float(x)) if float(x) < 0 else '{:.7f}'.format(float(x)))
```

```python
df_japon
```

Remplazo las columnas vacias con NaN

```python
# Reemplazar los valores vacíos con NaN
df_japon.replace('', np.nan, inplace=True)
```

Cambio Ｍ不明 a magnitud desconecida

```python
# Reemplazo los valores que estén en formato "Ｍ不明" por "Mag desconocida"
df_japon['mag'] = df_japon['mag'].replace('Ｍ不明', 'Mag desconocida')
```
